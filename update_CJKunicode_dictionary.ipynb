```python
import pandas as pd
```

```python
def generate_unicode_character_file(output_file='chinese_unicode.csv'):
    # Define all CJK Unicode blocks
    unicode_blocks = [
        ('CJK Unified Ideographs', 0x4E00, 0x9FFF),
        ('CJK Extension A', 0x3400, 0x4DBF),
        # For supplementary planes (requires Python 3.x)
        ('CJK Extension B', 0x20000, 0x2A6DF),
        ('CJK Extension C', 0x2A700, 0x2B73F),
        ('CJK Extension D', 0x2B740, 0x2B81F),
        ('CJK Extension E', 0x2B820, 0x2CEAF),
        ('CJK Extension F', 0x2CEB0, 0x2EBEF),
        ('CJK Extension G', 0x30000, 0x3134F),
        ('CJK Extension H', 0x31350, 0x323AF),
        ('CJK Compatibility', 0xF900, 0xFAFF),
        ('CJK Compatibility Supplement', 0x2F800, 0x2FA1F)
    ]
    
    data = []
    
    # Generate characters from all blocks
    for block_name, start, end in unicode_blocks:
        for code_point in range(start, end + 1):
            try:
                char = chr(code_point)
                unicode_hex = f'U+{code_point:04X}'
                data.append({
                    'character': char,
                    'unicode': unicode_hex,
                    'code_point': code_point,
                    'block': block_name
                })
            except:
                # Skip any issues with character generation
                pass
    
    # Create DataFrame and save to file
    df = pd.DataFrame(data)
    df_sorted = df.sort_values(by='code_point', ascending=True)
    df_sorted.to_csv(output_file, index=False)
    print(f"Generated {len(df_sorted)} Chinese characters to {output_file}")
    return df

# Create the file
chinese_chars_df = generate_unicode_character_file()
chinese_chars_df
```
